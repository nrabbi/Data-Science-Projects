```python
#import required libraries
import pandas as pd
import matplotlib.pyplot as plt

#reads file
episodes = pd.read_csv('datasets/office_episodes.csv')
```

```python
#matching specific value range to different colors in the plot
colors = []

for index, row in episodes.iterrows():
    if row['scaled_ratings'] < 0.25:
        colors.append('red')
    elif row['scaled_ratings'] >= 0.25 and row['scaled_ratings'] < 0.50:
        colors.append('orange')
    elif row['scaled_ratings'] >= 0.50 and row['scaled_ratings'] < 0.75:
        colors.append('lightgreen')
    else:
        colors.append('darkgreen')
```

```python
#apply the marker sizing based on guest appearance
marker = []

for index, row in episodes.iterrows():
    if row['has_guests'] == True:
        marker.append(250)
    else:
        marker.append(25)
```

```python
#find most watched episode
most_watched_episode = episodes['viewership_mil'].idxmax()

#find one of the guest star from the most watched episode
top_star = episodes.loc[most_watched_episode].guest_stars
top_star = top_star.split(",", 1)
top_star = top_star[0]
print(top_star) #output
```

```python
#plots data 
plt.rcParams['figure.figsize'] = [11, 7]
fig, ax = plt.subplots()
ax.scatter(episodes['episode_number'], episodes['viewership_mil'], c=colors, s=marker)
ax.set(xlabel='Episode Number', ylabel='Viewership (Millions)', title='Popularity, Quality, and Guest Appearances on the Office')
plt.show()
```
